```python
import pandas
data = pandas.read_csv('titanic.csv', index_col='PassengerId')
```

```python
data.head()
```

```python
full_names = data[['Name']].values
sexes = data[['Sex']].values
```

```python
first_names = []
for j in range(len(full_names)):
    if sexes[j] == 'female':
        tmp = full_names[j][0].split('.')[1]
        first_names.append(tmp.split('(')[0])
```

```python
first_names
```

```python
from collections import Counter
Counter(first_names).most_common()
```
